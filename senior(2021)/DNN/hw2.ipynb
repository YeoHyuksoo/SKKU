```python
import os
import random
import numpy as np
import torch
import torchvision.datasets as datasets
from torchvision import transforms
import matplotlib.pyplot as plt

transform = transforms.ToTensor()

train_dataset = datasets.MNIST(
    root='./',
    download=True,
    train=True,
    transform = transform
)

test_dataset = datasets.MNIST(
    root='./',
    download=False,
    train=False,
    transform = transform
)

from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split

train_loader = DataLoader(
    dataset=train_dataset,
    batch_size=200,
    num_workers=0,
)

test_loader = DataLoader(
    dataset=test_dataset,
    batch_size=200,
    num_workers=0,
)

import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(14*14*64, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = x.view(-1, 1, 28, 28)
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2, 2)
        x = x.view(-1, 14*14*64)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = Net()
model = model.to(device)

import torch.optim as optim
criterion = nn.NLLLoss()
optimizer = optim.Adam(model.parameters())

epochs = 30
x_axis = list(range(1, 31, 1))
y_axis = []
testacc_list = []
print("model 1\n")

for epoch in range(epochs):

    trainloss = 0
    train_total = 0
    train_correct = 0
    model.train()
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        trainloss += loss.item()

        _, predicted = outputs.max(1)
        train_total += labels.size(0)
        train_correct += predicted.eq(labels).sum().item()
    trainacc = train_correct / train_total

    print('[%d/%d] TrainLoss: %.3f, TrainAcc: %.3f' % (epoch+1, epochs, trainloss, trainacc * 100))
    y_axis.append(trainloss)

    model.eval()
    with torch.no_grad():
        correct = 0
        for x_test, y_test in test_loader:
            x_test, y_test = x_test.to(device), y_test.to(device)
            y_val = model(x_test)
            predicted = torch.max(y_val.data, 1)[1]
            correct += (predicted == y_test).sum()

    testacc = correct.item()/len(test_dataset)
    print('TestAcc: %.3f' % (testacc))
    testacc_list.append(testacc)

torch.save(model.state_dict(), './model1.pt')
print("save model1.pt\n")
new_model = Net()
new_model.load_state_dict(torch.load('./model1.pt'))

plt.plot(x_axis, y_axis)
plt.title('model 1 training loss vs epoch')
plt.xlabel('the number of epochs')
plt.ylabel('training loss')
plt.show()
plt.plot(x_axis, testacc_list)
plt.title('model 1 test accuracy vs epoch')
plt.xlabel('the number of epochs')
plt.ylabel('test accuracy')
plt.show()

class Net2(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.conv4 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(7*7*128, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = x.view(-1, 1, 28, 28)
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv3(x))
        x = F.relu(self.conv4(x))
        x = F.max_pool2d(x, 2, 2)
        x = x.view(-1, 7*7*128)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)

model2 = Net2()
model2 = model2.to(device)

optimizer = optim.Adam(model2.parameters())

y_axis = []
testacc_list = []
print("model 2\n")

for epoch in range(epochs):

    trainloss = 0
    train_total = 0
    train_correct = 0
    model2.train()
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        outputs = model2(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        trainloss += loss.item()

        _, predicted = outputs.max(1)
        train_total += labels.size(0)
        train_correct += predicted.eq(labels).sum().item()
    trainacc = train_correct / train_total

    print('[%d/%d] TrainLoss: %.3f, TrainAcc: %.3f' % (epoch+1, epochs, trainloss, trainacc * 100))
    y_axis.append(trainloss)

    model2.eval()
    with torch.no_grad():
        correct = 0
        for x_test, y_test in test_loader:
            x_test, y_test = x_test.to(device), y_test.to(device)
            y_val = model2(x_test)
            predicted = torch.max(y_val.data, 1)[1]
            correct += (predicted == y_test).sum()

    testacc = correct.item()/len(test_dataset)
    print('TestAcc: %.3f' % (testacc))
    testacc_list.append(testacc)

torch.save(model2.state_dict(), './model2.pt')
print('save model2.pt')
new_model2 = Net2()
new_model2.load_state_dict(torch.load('./model2.pt'))

plt.plot(x_axis, y_axis)
plt.title('model 2 training loss vs epoch')
plt.xlabel('the number of epochs')
plt.ylabel('training loss')
plt.show()
plt.plot(x_axis, testacc_list)
plt.title('model 2 test accuracy vs epoch')
plt.xlabel('the number of epochs')
plt.ylabel('test accuracy')
plt.show()

class Net3(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(14*14*32, 128)
        self.fc2 = nn.Linear(128, 10)
        self.residual_block = nn.Sequential(
            nn.Conv2d(32, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, stride=1, padding=1),
        )

    def forward(self, x):
        x = x.view(-1, 1, 28, 28)
        x = F.relu(self.conv1(x))
        x = self.residual_block(x)
        x = self.residual_block(x)
        x = F.max_pool2d(x, 2, 2)
        x = x.view(-1, 14*14*32)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)

model3 = Net3()
model3 = model3.to(device)

optimizer = optim.Adam(model3.parameters())

y_axis = []
testacc_list = []
print("model 3\n")

for epoch in range(epochs):

    trainloss = 0
    train_total = 0
    train_correct = 0
    model3.train()
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        outputs = model3(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        trainloss += loss.item()

        _, predicted = outputs.max(1)
        train_total += labels.size(0)
        train_correct += predicted.eq(labels).sum().item()
    trainacc = train_correct / train_total

    print('[%d/%d] TrainLoss: %.3f, TrainAcc: %.3f' % (epoch+1, epochs, trainloss, trainacc * 100))
    y_axis.append(trainloss)

    model3.eval()
    with torch.no_grad():
        correct = 0
        for x_test, y_test in test_loader:
            x_test, y_test = x_test.to(device), y_test.to(device)
            y_val = model3(x_test)
            predicted = torch.max(y_val.data, 1)[1]
            correct += (predicted == y_test).sum()

    testacc = correct.item()/len(test_dataset)
    print('TestAcc: %.3f' % (testacc))
    testacc_list.append(testacc)

torch.save(model3.state_dict(), './model3.pt')
print('save model3.pt')
new_model3 = Net3()
new_model3.load_state_dict(torch.load('./model3.pt'))

plt.plot(x_axis, y_axis)
plt.title('model 3 training loss vs epoch')
plt.xlabel('the number of epochs')
plt.ylabel('training loss')
plt.show()
plt.plot(x_axis, testacc_list)
plt.title('model 3 test accuracy vs epoch')
plt.xlabel('the number of epochs')
plt.ylabel('test accuracy')
plt.show()
```
